```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1.Read the dataset to Python environment;

```python
iris_data = pd.read_excel('iris.xls')
```

```python
iris_data = pd.DataFrame(iris_data)
```

```python
iris_data.head()
```

# 2. Display the columns in the dataset;

```python
iris_data.columns
```

# 3. Calculate the mean of each column of the dataset;

```python
value_SL = iris_data['SL']
mean_SL = np.mean(value_SL)
print('Mean of Sepal Length',mean_SL)
```

```python
value_SW = iris_data['SW']
mean_SW = np.mean(value_SW)
print('Mean of Sepal Width',mean_SW)
```

```python
value_PL = iris_data['PL']
mean_PL = np.mean(value_PL)
print('Mean of Petal Length ',mean_PL)
```

```python
value_PW = iris_data['PW']
mean_PW = np.mean(value_PW)
print('Mean of Petal Width',mean_PW)
```

# 4. Check for the null values present in the dataset;

```python
null_value = iris_data.isna().sum().sum()

null_data = iris_data[iris_data.isnull().any(axis=1)]

print('Total no.of Null value present in data is',null_value,'\n')

if null_value > 0:
    print('The null values present in the data are:\n')
    print(null_data)
else:
    print('Hence the data is good.') 
```

# 5. Perform meaningful visualizations using the dataset;

```python
plt.figure(figsize = (6,4))    
plt.hist(iris_data['SL'],rwidth = 0.8)
plt.title('Distribution of Sepal Length',fontsize = 16)
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
```

```python
plt.figure(figsize = (6,6) )    
plt.hist(iris_data['PL'],rwidth = 0.8)
plt.title('Distribution of Petal Length',fontsize = 16)
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
```

```python
plt.figure(figsize = (8,6) )    
plt.scatter(iris_data['SW'],iris_data['PW'],s = 10) 
plt.title('Relation of Sepal Width vs Petal Width', fontsize = 16)
plt.xlabel('Sepal Width')
plt.ylabel('Petal Width')
```

```python
plt.figure(figsize = (8,6))    
plt.scatter(iris_data['SL'],iris_data['PL'],s = 10) 
plt.title('Relation of Sepal Length vs Petal Length', fontsize = 16)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
plt.figure(figsize = (10,8))   
plt.subplot(2,1,1)
plt.scatter(iris_data['SL'],iris_data['PW'],s = 10) 
plt.title('Relation of Sepal Length vs Petal Width',fontsize = 16)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Width')

plt.subplot(2,1,2)
plt.scatter(iris_data['SW'],iris_data['PL'],s = 10) 
plt.title('Relation of Sepal Width vs Petal Length',fontsize = 16)
plt.xlabel('Sepal Width')
plt.ylabel('Petal Length')

plt.tight_layout()
```
